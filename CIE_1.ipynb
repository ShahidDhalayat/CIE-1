```python

import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/mtcars.csv')
plt.hist(data['mpg'],bins=10,edgecolor='k')
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('mtcars.csv')
plt.scatter(data['mpg'],data['wt'],color='blue')
plt.title('relation between weight of the car and mpg')
plt.xlabel('mpg')
plt.ylabel('wt')
plt.grid(True)
plt.show()
```
